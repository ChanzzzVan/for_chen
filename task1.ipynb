## Task 1-1

```python
# task 1-1
# 针对这个task我有2个想法：
# 想法1: 通过一轮对话，使用prompt设定身份和规则引导输出，这种方式比较“规矩”。
# 想法2：另一个是先在不设定角色的情况下先引导输出10个任意的未来科技，
#       再设定角色来连接这10个科技和未来方向盘功能的关系，这种方式更“天马行空”。
```

```python
#这里我用了moonshot的kimi，如果没有API可以用我提供的一个apikey
from openai import OpenAI
import re
import ast
client = OpenAI(
    ####################################
    api_key="",###这里需要填充API KEY####
    ###################################
    base_url="https://api.moonshot.cn/v1",
)
```

```python
# 想法1 

t1_sys_content11 = "你是专业的未来汽车趋势研究者，同时拥有汽车方向盘设计的丰富的理论基础和实践经验，富有想象力和创造力。" 
t1_user_content11 = """
## Goals（目标）:
生成一篇关于未来汽车方向盘功能趋势的文章，探讨未来汽车方向盘功能的趋势，考虑技术进步对方向盘设计的影响。

## Skills（技能）:
对汽车科技和趋势有深入了解。
熟悉当前方向盘功能的最新进展。
了解未来汽车设计的挑战与机遇。

## Workflows（工作流程）:
Step1.研究当前汽车方向盘功能现状。结果应是：>5个的功能总体方向，每方向有展开的>5个的具体功能。
Step2.了解未来趋势：发挥想象力展望未来可能影响方向盘功能的技术，思考这些技术将如何影响方向盘的设计和功能。
Step3.预测未来方向盘功能：根据step1.和step2.的研究，预测未来方向盘的功能和设计趋势。结果应是：给出未来方向盘功能的>5个的总体方向，每个方向有展开的>5个的具体功能，并作解释。

## Attention（注意事项）:
保持内容简洁明了，突出关键点。
使用真实的数据和案例支持观点。
请以正式的文章格式输出，包括适当的段落分隔和适当的技术术语解释。

## Output(输出)：
一篇关于未来汽车方向盘功能趋势的文章，字数在2000字左右。
"""
 
completion = client.chat.completions.create(
  model="moonshot-v1-8k",
  messages=[
    {"role": "system", "content": t1_sys_content11},
    {"role": "user", "content": t1_user_content11}
  ],
  temperature=0.3,
)
 
t1_output1 = completion.choices[0].message.content
print(t1_output1)
```

```python
#想法2

t1_user_content21 = "发挥想象力展望10个任意的未来科技"
t1_sys_content22 = "你是专业的未来科技趋势研究者，富有想象力和创造力，擅长通过天马行空的想象力创造两个概念之间的关联。"
t1_user_content22 = """
## Goals（目标）:
生成一篇关于未来汽车方向盘功能趋势的文章。

## Skills（技能）:
对未来科技有深入了解并充满想象。
熟悉当前汽车方向盘功能的最新进展。
了解未来汽车设计的挑战与机遇。

## Workflows（工作流程）:
发挥想象力将上轮对话的10个未来科技与方向盘功能设计相结合，给出未来方向盘功能的展望。

## Attention（注意事项）:
保持内容简洁明了，突出关键点。
请以正式的文章格式输出，包括适当的段落分隔和适当的技术术语解释。
"""
history = []
def chat(query, history):
    history += [{
        "role": "user", 
        "content": query
    }]
    completion = client.chat.completions.create(
        model="moonshot-v1-8k",
        messages=history,
        temperature=0.3,
    )
    result = completion.choices[0].message.content
    history += [{
        "role": "assistant",
        "content": result
    }]
    return result
 
chat(t1_user_content21, history)
history += [{
        "role": "system", 
        "content": t1_sys_content22
    }]
t1_output2 = chat(t1_user_content22, history)
print(t1_output2)
```

## Task 1-2

```python
# task 1-2
# step1: 提取生成文本中对于未来方向盘功能的预测
# step2：“和tesla model 3对比的结果”生成
# step3：评价生成的结果，最终结果是一个list，有三个分数，各表示准确性、全面性、语言和风格
```

```python
# step1: 提取生成文本中对于未来方向盘功能的预测
t2_user_content1 = f"""
# 工作流程：请根据以下文本提取出文中描述的“预测未来方向盘功能”的具体功能及其功能描述，并整理为json格式返回：

'''
{t1_output1}
'''

# 结果示例：
{{
  "具体功能1": "描述1",
  "具体功能2": "描述2"
}}

# 注意事项：
严格依据文本内容，禁止自行想象。
确保提取出文本中所有未来方向盘的功能，不要有遗漏，不要提取文本中其他部分的现状。
"""

completion = client.chat.completions.create(
  model="moonshot-v1-8k",
  messages=[
    {"role": "user", "content": t2_user_content1}
  ],
  temperature=1,
)
 
t2_output1 = completion.choices[0].message.content
print(t2_output1)
```

```python
# step2：和tesla model 3对比的结果生成
def gen_com(client,ur_model,prompt):
    completion = client.chat.completions.create(
      model=ur_model,
      messages=[
        {"role": "user", "content": prompt}
      ],
      temperature=0.3,
    )

    t2_output2 = completion.choices[0].message.content
    #print(t2_output2)
    return t2_output2

    
t2_user_content2 = f"""
我给出的文本中描述了某些方向盘功能和功能的描述，请你对比Tesla Model3方向盘功能和这些功能的差异。
'''
{t2_output1}
'''

#结果输出:
列出文本中的所有功能和功能描述，描述这些功能在Tesla Model3中是否有，如果有，有什么差异吗？
"""
ur_model= "moonshot-v1-8k"
t2_output2 = gen_com(client,ur_model,t2_user_content2)
print (t2_output2)
```

```python
# step3：评价生成的结果
t2_user_content3 =f"""
#任务
评价一份“方向盘功能文件和Tesla model3方向盘功能的差异对比”文章的质量。
该文章：
'''
{t2_output2}
'''

# 具体流程
Step1:你将在以下方面进行评价：
-1.信息准确性：评价该文章中涉及到tesla model3方向盘功能的情况是否符合真实情况，在下方5个分数中选择一个：
非常符合——5分；大多数符合，有一些内容不符合——4分；几乎一半符合，一半不符合——3分；大多数不符合——2分；几乎不符合——1分
-2.全面性：评价该文章是否覆盖了主要的未来可能的方向盘功能，在下方5个分数中选择一个：
几乎全部覆盖——5分；大多数覆盖——4分；一般覆盖——3分；小部份覆盖——2分；几乎没覆盖——1分
-3.语言和风格：文章的语言是否清晰、准确、专业，在下方5个分数中选择一个：
非常清晰、准确、专业——5分；比较清晰、准确、专业——4分；一般清晰、准确、专业——3分；不够清晰、准确、专业——2分；不清晰、准确、专业——1分

#输出：
最终结果以list形式输出，list中第一个值代表信息准确性的得分，第二个值代表全面性的得分，第三个值代表语言和风格的得分，放在生成文本的末尾。
输出示例：
[5,4,3]
"""
def gen_score(client,ur_model,prompt):
    t2_output3 = gen_com(client,ur_model,prompt)
    print(t2_output3)
    matches = re.findall(r'\[(.*?)\]', t2_output3)
# 输出匹配到的内容
    for match in matches:
        q = f'[{match}]' 
    list_score = ast.literal_eval(q)
    return list_score

list_score = gen_score(client,ur_model,t2_user_content3)
print(list_score)  
```
